#Matrix Multiplication Must Match

## Rule
```
The Inner Dimentions must Match and the Outer Dimentions will Determine the size of the Resultant Matrix.

```

*  Matrix_A ----> Size(3,4)
*  Matrix_B ----> Size(4,7)


```

                4 x 4
             ___________
            |           |
      ( 3 x 4 )   x   ( 4 x 7 )
        |___________________|
               3 x 7

```
For Matrix Multiplication we can use @ symbol or we can use matmul()
* matrix_A @ matrix_B
* torch.matmul(matrix_A, matrix_B)

````python
matrix_A = torch.rand(size=(3,4), dtype=torch.float32, device="cpu", requires_grad=False)
print(f"Below is the Matrix_A: \n{matrix_A}\n")
print(f"Size = {matrix_A.shape}")
print(f"Dimentions = {matrix_A.ndim}")
print(f"Total Elements = {matrix_A.numel()}")
````

````python
matrix_B = torch.rand(size=(4,7), dtype=torch.float32, device="cpu", requires_grad=False)
print(f"Below is the Matrix_A: \n{matrix_B}\n")
print(f"Size = {matrix_B.shape}")
print(f"Dimentions = {matrix_B.ndim}")
print(f"Total Elements = {matrix_B.numel()}")
````

````python
result = torch.matmul(matrix_A, matrix_B)
print(f"Below is the Resultant Matrix: \n{result}\n")
print(f"Size = {result.shape}")
print(f"Dimentions = {result.ndim}")
print(f"Total Elements = {result.numel()}")
````

# Reshaping, Stacking, Squeezing and Un-Squeezing Tensors

* Reshaping - Reshapes an Input Tensor to a defined Shape
* View - Returns a View of an input Tensor of certain shape but keep the same memory as the original Tensor
* Stacking - Combine Multiple Tensors on Top of each other Vertically (vstack) or Horizontally (hstack)
* Squeeze - Removes all 1 Dimentions from a given Tensor
* Un-Squeeze - Add 1 Dimention to the target Tensor
* Permute - Returns a View of the Input with Dimention Permuted (swapped) in a Certain way

````python
import torch

x = torch.randint(low=0, high=10, size=(1,12), dtype=torch.float32)
print(f"Below is the Tensor: \n{x}\n")
print(f"Size = {x.shape}")
print(f"Dimentions = {x.ndim}")
print(f"Total Elements = {x.numel()}")
````

````python
# Reshaping the Tensor x
x_reshaped = x.reshape(3,4)
print(f"Below is the Reshaped Tensor: \n{x_reshaped}\n")
print(f"Size = {x_reshaped.shape}")
print(f"Dimentions = {x_reshaped.ndim}")
print(f"Total Elements = {x_reshaped.numel()}")
````

````python
# Creating a View
z = x.view(1,12)
z

# Z is a view created as per x Because original Tensors share memory with views
# So, if they are sharing memory this means changing z would reflect the changes in z

# Changing z means changing x
print(x)
z[:, 3] = 25.6
z,x
````

````python
# Stacking
x_stack = torch.stack([x,x,x,x,x,x,x], dim=0)
x_stack
````

````python
# torch.squeeze() --> Removes all the single dimentions for a Targeted Tensor

# Creating a Tensor
tensor = torch.arange(1,10,1)
# Reshaping the Tensor
reshaped_tensor = tensor.reshape(1,9)



print(f"Tensor Before Squeezing: \n{reshaped_tensor}\n")
print(f"Size = {reshaped_tensor.shape}")
print(f"Dimentions = {reshaped_tensor.ndim}")
print(f"Total Elements = {reshaped_tensor.numel()}")

print(f"\n________________________________________________\n")
squzeed_tensor = reshaped_tensor.squeeze()
print(f"Tensor After Squeezing: \n{squzeed_tensor}\n")
print(f"Size = {squzeed_tensor.shape}")
print(f"Dimentions = {squzeed_tensor.ndim}")
print(f"Total Elements = {squzeed_tensor.numel()}")
````

````python
# torch.unsqueeze() --> Adds 1 in the shape of the tensor at a specific Dimention


print(f"Tensor Before Un-Squeezing: \n{squzeed_tensor}\n")
print(f"Size = {squzeed_tensor.shape}")
print(f"Dimentions = {squzeed_tensor.ndim}")
print(f"Total Elements = {squzeed_tensor.numel()}")

print(f"\n________________________________________________\n")
unsquzeed_tensor = squzeed_tensor.unsqueeze(dim=0)
print(f"Tensor After Squeezing: \n{unsquzeed_tensor}\n")
print(f"Size = {unsquzeed_tensor.shape}")
print(f"Dimentions = {unsquzeed_tensor.ndim}")
print(f"Total Elements = {unsquzeed_tensor.numel()}")
````

````python
# torch.permute --> Rearrange the Dimentions of a Targeted Tensor in a Specified Order

original_image = torch.rand(size=(244,244,3), dtype=torch.float32)
print(f"            Size of the Original Image")
print(f"Size = {original_image.shape}\n") # [height, width, color_channel]
print(f"\n________________________________________________\n")
permuted_image = original_image.permute(2,0,1) # shift axis 0->1, 1->2, 2->0
print(f"            Size of the Permuted Image")
print(f"Size = {original_image.shape}\n") # [color_channel, height, width]

original_image[0,0,0] = 23.00
print(original_image)
print("\n_______\n")
print(permuted_image)
````

#PyTorch Tensors and Numpy
Numpy is a popular Scientific Python Numerial Computing Laibrary
And because of this PyTorch has functionality to Interact with this

* Data in Numpy, want it in PyTorch --> `torch.from_numpy(ndarray)`
* PyTorch --> Numpy ---> `torch.Tensor.Numpy()`

````python
# Numpy Array to Tensors
import torch
import numpy as np

array = np.arange(1,12,2)

# Now we need to convert the Numpy Array to PyTorch Tensor
tensor = torch.from_numpy(array)

array, tensor
````

````python
# Verifing the data type of the Numpy Array and PyTorch Tensor
array.dtype, tensor.dtype
````

````python
# Changing the Numpy Array to check whether the changes reflect in the PyTorch Tensor
array = array + 1
array, tensor
````

````python
# Converting the PyTorch Tensor to Numpy Array
pytorch_tensor = torch.rand(size=(3,4))
numpy_array = pytorch_tensor.numpy()
pytorch_tensor, numpy_array
````

# Reproducbility (Trying to take Random out of Random)
In Short How Neural Network Learns:

`Start of Random Numbers --> Tensor Operations --> Update Random Number to Try and Make them of Data ---> again --> again --> again ... `

* To Reduce the concept of Randomness in Neural Network and PyTorch comes with the concept of Random Seed

Essentially what the Random Seed does is to "flavour" the concept of Randomness

````python
tensor_A = torch.rand(size=(3,4), dtype=torch.float32)
tensor_B = torch.rand(size=(3,4), dtype=torch.float32)
print(tensor_A)
print(tensor_B)
print(tensor_A == tensor_B)
````

````python
# As we have wittnised the Values in both the Tensors are not the same
# Now lets give it a try using RANDOM SEED

RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
tensor_C = torch.rand(size=(3,4), dtype=torch.float32)

torch.manual_seed(RANDOM_SEED)
tensor_D = torch.rand(size=(3,4), dtype=torch.float32)

print(tensor_C)
print(tensor_D)
print(tensor_C == tensor_D)
````
